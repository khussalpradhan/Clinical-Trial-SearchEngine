```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# -----------------------------
# INSERT YOUR RESULTS DICT HERE
# -----------------------------
results = {
    'mrr@10': 0.2553492063492063,
    'ndcg@3': 0.12715048917840702,
    'ndcg@5': 0.10580825286289278,
    'ndcg@10': 0.08999238967183021,
    'ndcg@20': 0.07836537754497994,
    'map@10': 0.005569999078078654,
    'precision@1': 0.16,
    'precision@3': 0.15555555555555556,
    'precision@5': 0.12,
    'precision@10': 0.10266666666666666,
    'precision@20': 0.08600000000000001,
    'recall@5': 0.006629491752884657,
    'recall@10': 0.009697031405852223,
    'recall@20': 0.015304280882272491,
    'recall@100': 0.015304280882272491,
    'hit_rate@1': 0.16,
    'hit_rate@5': 0.36,
    'hit_rate@10': 0.4533333333333333,
    'f1@5': 0.011983596511314316,
    'f1@10': 0.01662783552524296,
    'f1@20': 0.02387739806526702,
    'bpref': 0.01526851888358407
}

# Convert to DataFrame
df = pd.DataFrame(list(results.items()), columns=["Metric", "Value"])
df["Value"] = df["Value"].astype(float)


# ----------------------------------------------------
# 1. Print Markdown Table (copy/paste ready)
# ----------------------------------------------------
def print_markdown_table(df):
    print("| Metric | Value |")
    print("|--------|--------|")
    for _, row in df.iterrows():
        print(f"| **{row['Metric']}** | {row['Value']:.4f} |")


print("\n=== MARKDOWN TABLE ===\n")
print_markdown_table(df)


# ----------------------------------------------------
# 2. Save LaTeX Table
# ----------------------------------------------------
def save_latex(df, filename="evaluation_table.tex"):
    with open(filename, "w") as f:
        f.write("\\begin{table}[h]\n\\centering\n")
        f.write("\\begin{tabular}{l c}\n\\hline\n")
        f.write("\\textbf{Metric} & \\textbf{Value} \\\\\n\\hline\n")
        for _, row in df.iterrows():
            f.write(f"{row['Metric']} & {row['Value']:.4f} \\\\\n")
        f.write("\\hline\n\\end{tabular}\n")
        f.write("\\caption{Evaluation Metrics for Clinical Trial Retrieval System}\n")
        f.write("\\end{table}\n")

save_latex(df)
print("\nLaTeX table saved as evaluation_table.tex")


# ----------------------------------------------------
# 3. Chart helper
# ----------------------------------------------------
def plot_group(title, metrics):
    sub_df = df[df["Metric"].isin(metrics)]
    plt.figure(figsize=(8, 4))
    plt.bar(sub_df["Metric"], sub_df["Value"], color="steelblue")
    plt.title(title)
    plt.xticks(rotation=45, ha="right")
    plt.ylabel("Score")
    plt.tight_layout()
    plt.show()


# ----------------------------------------------------
# 4. Generate all Plots
# ----------------------------------------------------
plot_group("Ranking Metrics (MRR & nDCG)", 
           ["mrr@10", "ndcg@3", "ndcg@5", "ndcg@10", "ndcg@20"])

plot_group("Precision Metrics", 
           ["precision@1", "precision@3", "precision@5", "precision@10", "precision@20"])

plot_group("Recall Metrics", 
           ["recall@5", "recall@10", "recall@20", "recall@100"])

plot_group("Hit Rate Metrics",
           ["hit_rate@1", "hit_rate@5", "hit_rate@10"])

plot_group("F1 Metrics",
           ["f1@5", "f1@10", "f1@20"])

plot_group("Other Metrics",
           ["map@10", "bpref"])

print("\nCharts generated successfully.")
```
